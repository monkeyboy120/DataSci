```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

conn = sqlite3.connect('identifier.sqlite')
sql_query = "SELECT * FROM data"
df = pd.read_sql_query(sql_query, conn)
conn.close()

def plot_time_series(data, title):
    plt.figure(figsize=(12, 6))
    plt.plot(data)
    plt.title(title)
    plt.xlabel('Year')
    plt.ylabel(title)
    plt.grid(True)
    plt.show()

for column in df.columns.difference(['YEAR_MARK']):
    differenced_series = df[column].diff().dropna()
    plot_time_series(differenced_series, f'{column} (First Difference)')
```
